```python
#import the packages 
import pandas as pd
import numpy as np
```

```python
commerce_data=pd.read_csv('e_commerce.csv',sep=",")
print(commerce_data)
```

```python
commerce_data.isnull()
```

```python
commerce_data.isnull().sum()
```

```python
#Missing Percentage 
missingP_df = pd.DataFrame(commerce_data.isnull().sum())
missingP_df.columns = ['Missing Number']
missingP_df['Missing Percentage'] = (missingP_df['Missing Number']*100)/commerce_data.shape[0]
missingP_df
```

```python
commerce_data=commerce_data.drop(['Unnamed: 0'],axis=1)
commerce_data.head()
```

```python
commerce_data.isnull().sum()
```

```python
commerce_data.duplicated().sum()
```

```python
commerce_data.head(5)
```

```python
commerce_data['description'].dropna(axis=0)
commerce_data.head(10)
```

```python

commerce_data[['quantity', 'unit_price','number_of_visit','customer_age','rating_of_product']] =  commerce_data[['quantity', 'unit_price','number_of_visit','customer_age','rating_of_product']].fillna(commerce_data[['quantity', 'unit_price','number_of_visit','customer_age','rating_of_product']].median())
commerce_data.head(10)
```

```python
commerce_data.isnull().sum()
```

```python
commerce_data = commerce_data.dropna(subset=['description', 'country','invoice_date','customer_gender'])
```

```python
commerce_data.isnull().sum()
```

```python
commerce_data.head(20)
```

```python
commerce_data.head(10)
```

```python
commerce_data.head()
```

```python
commerce_data['invoice_date'] 
= np.where(commerce_data['invoice_date'].str.contains('/'), pd.to_datetime(commerce_data['invoice_date'])
.dt.strftime('%m/%d/%Y %H:%M'), 
pd.to_datetime(commerce_data['invoice_date'], dayfirst=True).dt.strftime('%m/%d/%Y %H:%M'))
```

```python
commerce_data.head()
```

```python
commerce_data.head(50)
```

```python
commerce_data.columns = commerce_data.columns.str.title()
commerce_data.columns
```

```python
commerce_data['Description']=commerce_data['Description'].str.title()
commerce_data['Customer_Gender']=commerce_data['Customer_Gender'].str.title()
```

```python
commerce_data.head()
```

```python
commerce_data.drop(commerce_data[commerce_data['Quantity'] >= 100].index, inplace = True),
commerce_data.drop(commerce_data[commerce_data['Unit_Price'] >= 50].index, inplace = True),
commerce_data.drop(commerce_data[commerce_data['Customer_Age'] >= 100].index, inplace = True),
commerce_data.drop(commerce_data[commerce_data['Country'] == 'Unspecified'].index, inplace = True),
commerce_data.tail(10)
```

```python
commerce_data.to_csv("e_commerce_finalclean.csv")
```
